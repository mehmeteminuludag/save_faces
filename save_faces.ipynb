```python
import tkinter as tk
import cv2
import numpy as np
import pandas as pd
from tkinter import Label,filedialog
from PIL import Image, ImageTk
import sys
import os
import face_recognition
import time
```

```python
class aplication:
    def __init__(self,root):
        self.root = root
        self.root.title("SaveFaces")
        self.persons=[]

        select_button = tk.Button(self.root,text="SelectVideo",command=self.select_video,width=10)
        select_button.place(x=10,y=10)
        process_button = tk.Button(self.root,text='StartProcess',command=self.main_process,width=10)
        process_button.place(x=110,y=10)
        quit_button = tk.Button(self.root,text='Quit',command=self.quit,width=10)
        quit_button.place(x=210,y=10)
        self.status_label = tk.Label(self.root)
        self.status_label.place(x=125,y=80)

    def quit(self):
         os._exit(0)
        
    def select_video(self):
        file_path = filedialog.askopenfilename(
        filetypes=[("Video Files", "*.mp4;*.avi;*.mov")],
        title="Select Video")
        self.video=cv2.VideoCapture(file_path)
        self.status_label.config(text="Video alındı")
    def main_process(self):
            cascade_path = r'C:\Users\Mehmet\anaconda3\envs\modelegitim\Lib\site-packages\data\haarcascade_frontalface_default.xml' #haarcascade path
            face_cascade = cv2.CascadeClassifier(cascade_path)
            face_list = [[],[]]
            person_count=0
            path_counter=0
            while True:
                back,frame=self.video.read()
                if back==True:
                        self.status_label.config(text="İşlem gerçekleştiriliyor")
                        faces = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5)
                        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                        for (x, y, w, h) in faces:
                            tmp_face=frame[y:y+h,x:x+w]
                            tmp_face=cv2.cvtColor(tmp_face, cv2.COLOR_BGR2RGB)
                            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)
                            tmp_face = face_recognition.face_encodings(tmp_face)
                            if tmp_face:
                                    tmp_face=tmp_face[0]
                                    matches = face_recognition.compare_faces(face_list[0], tmp_face)
                                    if len(face_list[0]) > 0 and any(matches):
                                        index = matches.index(True)
                                        face_list[1][index]+=1
                                        path_counter=path_counter
                                        photo_counter=face_list[path_counter]
                                    else:
                                        person_count+=1
                                        path_counter=0
                                        path=r'C:\Users\Mehmet\Desktop\save_faces\persons\\'+str(person_count) #file path
                                        os.makedirs(path, exist_ok=True)
                                        face_list[0].append(tmp_face)
                                        face_list[1].append(0)
                                        index=0
                                    face_file_path=r'C:\Users\Mehmet\Desktop\save_faces\persons\\'+str(person_count)
                                    face_path=face_file_path+'\photo'+str(face_list[1][index])+'.jpg' #new image path
                                    cv2.imwrite(face_path,frame)
                        time.sleep(0.5)
                else :
                    self.status_label.config(text="İşlem bitti")
                    break
                
```

```python
window=tk.Tk()
window.geometry("300x200")
ap1=aplication(window)
window.mainloop()
                
```
